```python
directory = 'D:\\BIT_study_related\\Thesis\\Data_Monthly\\OCO-2\\netcdf'

def conv_date(d):
    return datetime.strptime(str(d), '%Y%m%d%H%M%S%f')
for files in os.listdir(directory):
    y = ''
    m = ''
    d = ''
    data = Dataset(files)
    df_xco2= pd.DataFrame(columns=["DateTime", "Latitude", "Longitude", "xco2", "xco2_quality_flag","pressure_levels"])
    df_xco2["xco2"]= data["xco2"][:]
    df_xco2["pressure_levels"]= data["pressure_levels"][:]
    df_xco2["DateTime"]= data["sounding_id"][:]
    df_xco2["Latitude"]= data["latitude"][:]
    df_xco2["Longitude"]= data["longitude"][:]
    df_xco2["xco2_quality_flag"]= data["xco2_quality_flag"][:]
    df_xco2["DateTime"]= df_xco2["DateTime"].apply(conv_date)
    df_xco2["DateTime"]= pd.to_datetime(df_xco2["DateTime"])
    df_xco2["Year"]= df_xco2["DateTime"].dt.year
    df_xco2["Month"]= df_xco2["DateTime"].dt.month
    df_xco2["Day"]= df_xco2["DateTime"].dt.day
    y = str(df_xco2.Year[0])
    if y == '2023':
        m = str(df_xco2.Month[0])
        d = str(df_xco2.Day[0])
        ymd = y+str('_')+m+str('_')+d
        df_xco2.to_csv('D://BIT_study_related//Thesis//Data_Monthly//OCO-2//2023//OCO-2_'+ymd+'.csv')
```
